```python
import matplotlib.pyplot as plt
sns.countplot(x="departure_time",data=df)
plt.xticks(rotation=90)
plt.title("departure_time")
plt.show()
sns.countplot(x="arrival_time",data=df)
plt.title("arrival_time")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x="departure_time",y="price",data=df,palette="plasma")
plt.xlabel("departure_time")
plt.ylabel("ticket price")
plt.xticks(rotation=90)
plt.show()
#plt.subplots(212)
plt.figure(figsize=(5,5))
sns.barplot(x="arrival_time",y="price",data=df,color="c")
plt.xlabel("arrival_time")
plt.ylabel("ticket price")
plt.xticks(rotation=90)
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x="source_city",y="price",data=df,palette="plasma")
plt.xlabel("source_city")
plt.ylabel("ticket price")
plt.xticks(rotation=90)
plt.show()
#plt.subplots(212)
plt.figure(figsize=(5,5))
sns.barplot(x="destination_city",y="price",data=df,palette="plasma")
plt.xlabel("destination_city")
plt.ylabel("ticket price")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x="class",y="price",data=df,palette="plasma")
plt.xlabel("class")
plt.ylabel("ticket price")
plt.title("economy vs business ticket price")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x="airline",y="price",data=df,palette="plasma")
plt.xlabel("airline")
plt.ylabel("ticket price")
plt.title("airline selling price")
plt.xticks(rotation=90)
plt.show()
```
